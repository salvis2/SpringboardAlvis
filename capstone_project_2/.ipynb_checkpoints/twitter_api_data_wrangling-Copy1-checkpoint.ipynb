### Time Sensitivity

Try to get 25k - 50k tweets. I will try to get 20k tweets from each block (nintendo / pokemon and before / after) to account for duplicate tweets.

I can only pull tweets from the last week ish. For the tweets that encapsulate Nintendo's activity pre-announcement, I have less time to get them than the tweets after the announcement. Getting followers / friendships can happen at any time.

```python
# This block should search for any recent tweets involving pokemon
params = {
    # General search term
    'q': 'pokemon',
    
    # Others
    'lang': 'en',
    'result_type': 'recent',
    'count': '100',
    'include_entities': 'false',
    'max_id': '1202121092154417152'
}
```

```python
pokemon_tweets_list_after = []
poke_tweet_count_after = []

while (len(pokemon_tweets_list_after) < 200):
    # Get the tweets, put their df in the list
    tweets = requests.get(twitter_url, params=params, auth=auth)
    df = json_normalize(tweets.json()['statuses'])
    pokemon_tweets_list_after.append(df)
    poke_tweet_count_after.append(len(df))
    
    # Set up max ID for the next call
    # Find the smallest ID value
    # ID is column 6
    current_max = df.sort_values('id', ascending=True).iloc[0, 6] - 1 
    params['max_id'] = str(current_max) if current_max < int(params['max_id']) else params['max_id']
    
    # I can do 180 requests per 15 minutes
    # This is the same as a request every 5 seconds
    # I will go a little slower than that
    time.sleep(6)

len(pokemon_tweets_list_after)
```
